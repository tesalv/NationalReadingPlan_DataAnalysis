```python

filenames = glob.glob(path + "\*.csv")

# for loop to iterate all csv files

authors_list=[]
for file in filenames:
    data=pd.read_csv(file,encoding='latin1')
    data.drop_duplicates(subset="1st_author")
    df_author=list(data['1st_author'])

    for author in df_author:
        if author not in authors_list:
            authors_list.append(author)
        else:
            pass

#print(authors_list)
```

```python
unique_authors=pd.DataFrame(authors_list,columns=["author"])
unique_authors.head()
```

```python
unique_authors.to_csv(r"datasets\unique_authors.csv",encoding='latin1')
```
