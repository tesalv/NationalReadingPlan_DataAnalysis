```python
from bs4 import BeautifulSoup
import requests
```

```python
base_url= "https://www.google.com/search?q="

authours_names=["Mia Couto","George Orwell","Neil Gaiman","Alice Hoffman ","Pedro Strecht","Lucy Maud Montgomery","Antoine de Saint-Exupéry "]

author=authours_names[0]

author_split=author.split()
#print(author_split)
#print(len(author_split))
author_url=""
i= 0
for i in range(0,len(author_split)):
    author_url=author_url + author_split[i] + "+"
    i= i+1

final_url=base_url+author_url+"nationality"
final_url
```

```python
response =  requests.get(final_url)
response.status_code
```

```python

soup = BeautifulSoup(response.content)
soup
```
