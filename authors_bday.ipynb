```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

import time
import random
```

```python
data=pd.read_csv("datasets\PNL_15 - 18 anos_2022 - 1.o Sem..csv",encoding = 'latin1')
data.head()
```

```python
base_url= "https://en.wikipedia.org/wiki/"

authours_names=data["1st_author"]

bday=[]

#author=authours_names[-1]
for author in authours_names:
    author_split=author.split()
    #print(author_split)
    #print(len(author_split))
    author_url=""
    i= 0
    for i in range(0,len(author_split)):
        author_url=author_url + author_split[i] + "_"
        i= i+1

    final_url=base_url+author_url
    final_url
    print(final_url)
    response =  requests.get(final_url)
    response.status_code

    soup = BeautifulSoup(response.content)
    #print(soup.head())
    try:
        birthday = soup.find("span",attrs={"class":"bday"}).text
        bday.append(birthday)
    except:
        bday.append("0")
        continue
    time.sleep(random.randint(0,5))

#short_data=data[0:5]
data["bday"] =np.array(bday)

data.head()
```
