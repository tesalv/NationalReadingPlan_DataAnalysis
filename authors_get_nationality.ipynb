# function

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://www.google.com')
```

```python
def get_author_nationality(authorli):
    author_data = []
    base_url= "https://www.google.com/search?q="
    for author in authorli:
        author_split=author.split()
        author_url=""
        author_url= '+'.join([i for i in author_split])
        final_url=base_url+author_url+"+nationality"
        print(final_url)

        driver.get(final_url)
        time.sleep(3)
        html = driver.page_source
        soup = BeautifulSoup(html)
        time.sleep(3)
        try:
            nat = soup.find('div', attrs= {'data-tts': 'answers'}).text
            
            author_data.append({'name': author, 'nationality': nat})

        except:
            try:
                desc = soup.find('div', attrs={'data-attrid': 'wa:/description'})
                nat = desc.find('b').text
                author_data.append({'name': author, 'nationality': nat})
            except:
                author_data.append({'name': author, 'nationality': 'none'})
    
    return author_data
```
